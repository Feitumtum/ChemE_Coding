```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def y(t):
    return 3/10*(np.exp(-t) + 1/3*np.sin(3*t) - np.cos(3*t))
```

```python
t = np.linspace(-2, 40, 500)
t = np.linspace(0, 40, 1000)

plt.plot(t,y(t))
plt.title( "Y(t) vs t")
plt.ylabel("Y(t)")
plt.xlabel("t")
plt.show()
```
